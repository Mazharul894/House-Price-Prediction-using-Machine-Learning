```python
#import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#Import Data set
import pandas as pd
data_url = "http://lib.stat.cmu.edu/datasets/boston"
# Load the raw dataset
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
# Combine rows into a single row per observation
data = []
for i in range(0, len(raw_df), 2):
    # Combine row i and row i+1, dropping NaN values (to avoid trailing spaces)
    combined_row = list(raw_df.iloc[i].dropna()) + list(raw_df.iloc[i + 1].dropna())
    data.append(combined_row)
# Ensure every row has exactly 14 values
data = [row for row in data if len(row) == 14]
# Define column names
column_names = [ "CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE","DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "PRICE"]
# Create the final DataFrame
housing_price_df = pd.DataFrame(data, columns=column_names)
# Display the first few rows
housing_price_df.head()
```

```python
housing_price_df.isnull().sum()
```

```python
housing_price_df.describe()
```

```python
#Fined the corelation between the data
corr = housing_price_df.corr()
corr
```

```python
#create plot of corr
plt.figure(figsize=(10,10))
sns.heatmap(corr, annot=True, fmt='.1f', cmap="Blues")
```

```python
##spliting the dataset
X = housing_price_df.drop(['PRICE'], axis=1)
y = housing_price_df['PRICE']
```

```python
#Spliting the data into training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
#Model training
model = XGBRegressor()
model.fit(X_train, y_train)
```

```python
#preduction on training data
training_data_prediction = model.predict(X_train)
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(y_train, training_data_prediction)
print("R squared error : ", score_1)
#Mean Absolute Error
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)
print("Mean Absolute Error : ", score_2)
```

```python
#preduction on test data
test_data_prediction = model.predict(X_test)
print(test_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(y_test, test_data_prediction)
print("R squared error : ", score_1)
#Mean Absolute Error
score_2 = metrics.mean_absolute_error(y_test, test_data_prediction)
print("Mean Absolute Error : ", score_2)
```

```python
plt.scatter(y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
